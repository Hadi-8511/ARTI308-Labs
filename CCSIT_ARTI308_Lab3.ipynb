# Lab 3: Exploratory Data Analysis (EDA)

```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set()
```

```python
# Load Dataset
df = pd.read_csv("Student_performance_Dataset.csv")

# Display first 5 rows
df.head()
```

```python
# Check all values
print(df.isna())
```

```python
print(df.isna().sum())
```

### Check duplicate rows

```python
# checking duplicate rows
df.duplicated()[df.duplicated()==True]
```

### No. of rows and columns

```python
# finding number of rows and columns

print("Shape (rows, columns): ", df.shape,"\n")

print("number of rows: ", df.shape[0])
print("number of columns: ", df.shape[1])
```

### Data type of columns

```python
# viewing the data types of columns
df.dtypes
```

```python
df['GradeClass'] = pd.to_numeric(df['GradeClass'].astype(int))
df.dtypes
```

### Descriptive summary Statistics

```python
# Statistical summary
df.describe(include='all')
```

### Univariate Analysis

```python
plt.figure(figsize=(8,5))
sns.countplot(x='GradeClass', data=df)
plt.title('Distribution of GradeClass')
plt.ylabel('Number of Students')
plt.show()
```

## Bivariate Analysis

### GPA by GradeClass


```python
grade_gpa_summary = df.groupby('GradeClass')['GPA'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x='GradeClass', y='GPA', data=df,errorbar=None)
plt.title("Impact of GPA on Grade Category", fontsize=14)
plt.ylabel("Average GPA")
plt.show()

grade_gpa_summary 
```

### Absences by GradeClass

```python
absences_summary = df.groupby('GradeClass')['Absences'].mean().sort_values(ascending=True)

plt.figure(figsize=(10, 5))
sns.barplot(x='GradeClass', y='Absences', data=df,errorbar=None)
plt.title("Absences by Grade Class", fontsize=14)
plt.ylabel("Average Number of Absences")
plt.show()

absences_summary
```

### Study Time by GradeClass

```python
study_time_summary = df.groupby('GradeClass')['StudyTimeWeekly'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='GradeClass', y='StudyTimeWeekly', data=df,errorbar=None)
plt.title("Study Time Weekly by Grade Class", fontsize=14)
plt.ylabel("Average Weekly Study Time (Hours)")
plt.show()

study_time_summary
```

### Study Time vs GPA


```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='StudyTimeWeekly', y='GPA', data=df)
plt.title('Study Time vs GPA')
plt.show()
```

### Correlation Matrix

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[['StudyTimeWeekly', 'GPA','ParentalSupport', 'Absences']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

- Correlation close to 1 = strong positive relationship
- Close to 0 = weak relationship
